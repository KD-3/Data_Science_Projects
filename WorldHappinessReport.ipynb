# World Happiness Report

**Importing World Happiness Report data for year 2015, 2016 and 2017**

```
import pandas as pd

# data2015 = pd.read_csv("2015.csv").sort_values("Country").sort_values("Country")
data_2015 = pd.read_csv("2015.csv").sort_values("Country")
data_2015 = data_2015.set_index(data_2015["Country"])
data_2016 = pd.read_csv("2016.csv").sort_values("Country")
data_2016 = data_2016.set_index(data_2016["Country"])
data_2017 = pd.read_csv("2017.csv").sort_values("Country")
data_2017 = data_2017.set_index(data_2017["Country"])
```

**Countries with a happiness score of less than 5.0 in 2015**

```
data_2015[['Happiness Score']][data_2015["Happiness Score"] < 5]
```

**Countries with a happiness score of less than 5.0 in 2016**

```
data_2016[['Happiness Score']][data_2016["Happiness Score"] < 5]
```

**Countries with a happiness score of less than 5.0 in 2015**

```
data_2017[['Happiness.Score']][data_2017["Happiness.Score"] < 5]
```

 **The unhappiest country in Sub-Saharan Africa**

```
data_sahara_2015 = data_2015[data_2015["Region"] == "Sub-Saharan Africa"]
data_sahara_2016 = data_2016[data_2016["Region"] == "Sub-Saharan Africa"]
```

```
data_sahara_2015[data_sahara_2015["Happiness Score"] == data_sahara_2015["Happiness Score"].min()]
```

```
data_sahara_2016[data_sahara_2016["Happiness Score"] == data_sahara_2016["Happiness Score"].min()]
```

**Calculating the average happiness for all the countries for the years 2015-2017**

```
data_2015["Average Happiness"] = (data_2015["Happiness Score"]+data_2016["Happiness Score"]+data_2017["Happiness.Score"])/3
```

**The Unhappiest country in each region for the years 2015-2017**



```
data_2015[['Region', 'Average Happiness']].sort_values(by = 'Average Happiness').groupby('Region').head(1)
```

**The Happiest country in each region for the years 2015-2017**



```
data_2015[['Region', 'Average Happiness']].sort_values(by = 'Average Happiness').dropna().groupby('Region').tail(1)
```

**Countries that became unhappy between 2015 to 2017**

```
data_unhappy_2015to2017 = data_2015['Happiness Score'] - data_2017['Happiness.Score']
data_2015['unhappy_meter2015'] = data_unhappy_2015to2017
data_unhappy_2016to2017 = data_2016['Happiness Score'] - data_2017['Happiness.Score']
data_2015['unhappy_meter2016'] = data_unhappy_2016to2017
data_2016['Happiness Score']
```

```
data_2015[['Country', 'unhappy_meter2015']][data_2015['Happiness Score'] - data_2017['Happiness.Score'] > 0]
```

**Country whose happiness decreased by the most amount**

```
data_2015[['Country', 'unhappy_meter2015']][data_2015['Happiness Score'] - data_2017['Happiness.Score'] > 0].max()
```
